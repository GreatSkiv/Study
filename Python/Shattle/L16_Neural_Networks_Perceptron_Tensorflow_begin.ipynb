#### Make Class as sklearn with  methods fit and predict for task classification

```python
###  Модель работает только для бинарной классификации (-1,1)
class Perceptron:
    def __init__(self, eta=0.01, n=10):
        self.eta = eta
        self.n = n
    # net_input() - действие происходящее внутри перцептрона
    # реализация уравнения гиперплоскости x[1]*w[1]+x[1]*w[1]+...+x[n]*w[n] +b
    def net_input(self, x):
        # @-> умножение матриц
        return x @ self.w + self.b
    
    def predict(self, x):
        return np.where(self.net_input(x) > 0, 1, -1)
       
    def fit(self, x, y):
        # x-вектор признаков(матрица) массив numpy ,y- классы(вектор ответов) из -1 или 1
        # w масссив весов,  количество, столько сколько и х
        self.w = np.random.normal(size=x.shape[1])
        self.b = 0
        self.errors = []
        # Далее процесс обучения в количестве  итераций (n)
        #Обучение по паре вход/выход значений. Расчитываем угадали выход или нет и двигать вес
        # x_- строка набор признаков, y_ - конкретный результат
        for i in range(self.n):
            errors = 0
            for x_, y_ in zip(x, y):
                update = y_ - self.predict(x_)
                self.w += self.eta*(update * x_)
                self.b += self.eta*update
                #print(self.w, self.b, y_, self.predict(x_))
                errors += y_ != self.predict(x_)
                self.errors.append(errors)
```

#### Проверку работы разработанной модели производим на встроенном дата сете iris

```python
from sklearn.datasets import load_iris
iris = load_iris()
```

#### Для проверки берем 2 признака из даты и 2 группы(1 и 2-ю) из таргета(100 элементов)

```python
data = iris.data[:100, 2:]
target = iris.target[:100]
```

#### Так как на выходе пепцетрона могут быть только -1 или 1 преобразуем таргет

```python

target = np.where(target == 0, -1, 1)
target
```

```python
p = Perceptron()
p.fit(data, target)
p.errors 
```

```python
import matplotlib.pyplot as plt
xmin, xmax = data[:, 0].min() - 1, data[:, 0].max() + 1
ymin, ymax = data[:, 1].min() - 1, data[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(xmin, xmax, 0.01), np.arange(ymin, ymax, 0.01)) # задает сетку на плоскости 
z = np.c_[xx.ravel(), yy.ravel()] #ravel()  Делает массив плоским, np.c_ -concatenation 
Z = np.zeros_like(xx) #Возвращает массив нулей той же формы и типа, что и заданный массив.
Z = Z.reshape(1, -1) #Дает новую форму массиву без изменения его данных.
Z = Z[0]

for i, d in enumerate(z):#создает объект, который генерирует кортежи, состоящие из двух элементов - индекса элемента и самого элемента
    Z[i] = p.predict(d)
Z = Z.reshape(xx.shape)
plt.pcolormesh(xx, yy, Z) #Создайте псевдоколорный график с нерегулярной прямоугольной сеткой.
plt.scatter(data[:, 0][:50], data[:, 1][:50])
plt.scatter(data[:, 0][50:], data[:, 1][50:])
plt.show()
```
